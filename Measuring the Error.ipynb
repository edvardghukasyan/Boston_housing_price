```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=[['Linear Regression',3.703897,0.732038,0.789922,0.701389],['Ridge Regression',2.360645,0.885407,0.914665,0.828044],['Lasso Regression',2.533564,0.906332,0.901706,0.837923],['Support Vector Regression',2.399574,0.886519,0.911828,0.817580],['Decision Tree Regression',2.886515,1.000000,0.872412,0.762822]]
Data=pd.DataFrame(data,columns=['Model','RMSE','R2_Score_Training','R2_Score_Test','Cross_Validation'])
Data
```

```python
plt.barh(Data['Model'], Data['RMSE'],color=["y","b","r","c","g"])
plt.xlabel('RMSE')
plt.ylabel('Model')
plt.show()
```

```python
plt.barh(Data['Model'], Data['R2_Score_Training'])
plt.xlabel('R2_Score_Training')
plt.ylabel('Model')
plt.show()
plt.barh(Data['Model'], Data['R2_Score_Test'])
plt.xlabel('R2_Score_Test	')
plt.ylabel('Model')
plt.show()
```

```python
plt.barh(Data['Model'], Data['Cross_Validation'])
plt.xlabel('Cross_Validation')
plt.ylabel('Model')
plt.show()
```
